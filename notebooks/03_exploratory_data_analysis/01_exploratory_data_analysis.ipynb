#### Step 4: Analysis of Experience and Job Satisfaction


Analyze the relationship between years of professional coding experience (`YearsCodePro`) and job satisfaction (`JobSat`). Summarize `YearsCodePro` and calculate median satisfaction scores based on experience ranges.

- Create experience ranges for `YearsCodePro` (e.g., `0-5`, `5-10`, `10-20`, `>20` years).

- Calculate the median `JobSat` for each range.

- Visualize the relationship using a bar plot or similar visualization.


```python
## Write your code here
# Estadísticas descriptivas de YearsCodePro
df["YearsCodePro"].describe()
```

```python
# Revisar los valores no numéricos más frecuentes
df["YearsCodePro"].value_counts(dropna=False).head(20)
```

```python
import pandas as pd

# Normalizar casos especiales
df["YearsCodePro"] = df["YearsCodePro"].replace({
    "Less than 1 year": 0.5,
    "More than 50 years": 51
})

# Convertir cualquier valor restante a numérico
# Los textos no reconocidos se transformarán en NaN
df["YearsCodePro"] = pd.to_numeric(
    df["YearsCodePro"],
    errors="coerce"
)

# Verificar el resultado
print(df["YearsCodePro"].dtype)
print(df["YearsCodePro"].isna().sum())
```

```python
# Definir límites y etiquetas
bins = [0, 5, 10, 20, float("inf")]
labels = ["0-5", "5-10", "10-20", ">20"]

# Crear rangos de experiencia
df["ExperienceRange"] = pd.cut(
    df["YearsCodePro"],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)

# Revisar la cantidad de observaciones por rango
df["ExperienceRange"].value_counts(dropna=False).sort_index()
```

```python
# Mediana de satisfacción por rango de experiencia
median_jobsat = (
    df.groupby("ExperienceRange", observed=True)["JobSat"]
      .median()
      .reset_index()
)

median_jobsat
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

plt.bar(
    median_jobsat["ExperienceRange"].astype(str),
    median_jobsat["JobSat"]
)

plt.title("Median Job Satisfaction by Professional Coding Experience")
plt.xlabel("Years of Professional Coding Experience")
plt.ylabel("Median Job Satisfaction")

plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

#### Step 6: Analyzing Remote Work Preferences by Job Role


Analyze trends in remote work based on job roles. Use the `RemoteWork` and `Employment` columns to explore preferences and examine if specific job roles prefer remote work more than others.

- Use a count plot to show remote work distribution.

- Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.


```python
## Write your code here

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9, 5))

sns.countplot(
    data=df,
    x="RemoteWork",
    order=df["RemoteWork"].value_counts().index
)

plt.title("Distribution of Remote Work")
plt.xlabel("Remote Work Type")
plt.ylabel("Number of Respondents")

plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
remote_employment_table = pd.crosstab(
    df["Employment"],
    df["RemoteWork"]
)

remote_employment_table
```

```python
remote_employment_pct = pd.crosstab(
    df["Employment"],
    df["RemoteWork"],
    normalize="index"
).mul(100).round(2)

remote_employment_pct
```

```python
# Seleccionar los 10 tipos de empleo más frecuentes
top_employment = df["Employment"].value_counts().head(10).index

df_employment_top = df[
    df["Employment"].isin(top_employment)
]

plt.figure(figsize=(12, 7))

sns.countplot(
    data=df_employment_top,
    y="Employment",
    hue="RemoteWork",
    order=top_employment
)

plt.title("Remote Work Distribution by Employment Type")
plt.xlabel("Number of Respondents")
plt.ylabel("Employment Type")
plt.legend(title="Remote Work")

plt.tight_layout()
plt.show()
```

#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
## Write your code here
# Top 10 países con mayor número de respuestas
top_countries = df["Country"].value_counts().head(10).index

df_region = df[
    df["Country"].isin(top_countries)
][["Country", "LanguageHaveWorkedWith"]].dropna()
```

```python
languages = (
    df_region
    .assign(
        LanguageHaveWorkedWith=df_region["LanguageHaveWorkedWith"].str.split(";")
    )
    .explode("LanguageHaveWorkedWith")
)

languages["LanguageHaveWorkedWith"] = (
    languages["LanguageHaveWorkedWith"]
    .str.strip()
)
```

```python
language_counts = (
    languages
    .groupby(["Country", "LanguageHaveWorkedWith"])
    .size()
    .reset_index(name="Count")
)

language_counts.head()
```

```python
top_languages = (
    language_counts
    .sort_values(["Country", "Count"], ascending=[True, False])
    .groupby("Country")
    .head(10)
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))

sns.barplot(
    data=top_languages,
    x="Count",
    y="LanguageHaveWorkedWith",
    hue="Country"
)

plt.title("Top Programming Languages by Country")
plt.xlabel("Number of Developers")
plt.ylabel("Programming Language")

plt.tight_layout()
plt.show()
```

```python
heatmap_data = (
    language_counts
    .pivot(
        index="LanguageHaveWorkedWith",
        columns="Country",
        values="Count"
    )
    .fillna(0)
)

plt.figure(figsize=(12, 10))

sns.heatmap(
    heatmap_data,
    cmap="Blues"
)

plt.title("Programming Language Popularity by Country")
plt.xlabel("Country")
plt.ylabel("Programming Language")

plt.tight_layout()
plt.show()
```

#### Step 8: Correlation Between Experience and Satisfaction


Examine how years of experience (`YearsCodePro`) correlate with job satisfaction (`JobSatPoints_1`). Use a scatter plot to visualize this relationship.


```python
## Write your code here

import pandas as pd

df["YearsCodePro"] = pd.to_numeric(
    df["YearsCodePro"],
    errors="coerce"
)

df["JobSatPoints_1"] = pd.to_numeric(
    df["JobSatPoints_1"],
    errors="coerce"
)

experience_satisfaction = (
    df[["YearsCodePro", "JobSatPoints_1"]]
    .dropna()
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9, 6))

sns.scatterplot(
    data=experience_satisfaction,
    x="YearsCodePro",
    y="JobSatPoints_1",
    alpha=0.3
)

plt.title("Professional Experience vs. Job Satisfaction")
plt.xlabel("Years of Professional Coding Experience")
plt.ylabel("Job Satisfaction Points")

plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
pearson_corr = experience_satisfaction[
    ["YearsCodePro", "JobSatPoints_1"]
].corr(method="pearson").iloc[0, 1]

spearman_corr = experience_satisfaction[
    ["YearsCodePro", "JobSatPoints_1"]
].corr(method="spearman").iloc[0, 1]

print(f"Pearson correlation: {pearson_corr:.3f}")
print(f"Spearman correlation: {spearman_corr:.3f}")
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
## Write your code here

education_employment = (
    df[["EdLevel", "Employment"]]
    .dropna()
    .assign(Employment=lambda x: x["Employment"].str.split(";"))
    .explode("Employment")
)

education_employment["Employment"] = (
    education_employment["Employment"].str.strip()
)
```

```python
education_employment_pct = pd.crosstab(
    education_employment["EdLevel"],
    education_employment["Employment"],
    normalize="index"
).mul(100).round(2)

education_employment_pct
```

```python
top_employment = (
    education_employment["Employment"]
    .value_counts()
    .head(6)
    .index
)

education_employment_top = education_employment[
    education_employment["Employment"].isin(top_employment)
]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))

sns.countplot(
    data=education_employment_top,
    y="EdLevel",
    hue="Employment",
    order=education_employment_top["EdLevel"].value_counts().index
)

plt.title("Employment Type by Educational Background")
plt.xlabel("Number of Respondents")
plt.ylabel("Educational Level")
plt.legend(
    title="Employment Type",
    bbox_to_anchor=(1.05, 1),
    loc="upper left"
)

plt.tight_layout()
plt.show()
```
