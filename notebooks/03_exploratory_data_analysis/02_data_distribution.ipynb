### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
# Missing values by column
missing_values = df.isnull().sum()

missing_values[missing_values > 0].sort_values(ascending=False)
```

```python
print(f"Total missing values: {df.isnull().sum().sum()}")
```

```python
numeric_cols = df.select_dtypes(include="number").columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
categorical_cols = df.select_dtypes(include="object").columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
import pandas as pd

# Languages currently used
worked = (
    df["LanguageHaveWorkedWith"]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
    .value_counts()
)

# Languages developers want to use
wanted = (
    df["LanguageWantToWorkWith"]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
    .value_counts()
)
```

```python
language_comparison = pd.concat(
    [worked, wanted],
    axis=1,
    keys=["WorkedWith", "WantToWorkWith"]
).fillna(0)

language_comparison = language_comparison.astype(int)
```

```python
top_languages = (
    language_comparison
    .sum(axis=1)
    .sort_values(ascending=False)
    .head(10)
    .index
)

plot_data = (
    language_comparison
    .loc[top_languages]
    .reset_index()
    .rename(columns={"index": "Language"})
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plot_data = plot_data.melt(
    id_vars="Language",
    var_name="Category",
    value_name="Count"
)

plt.figure(figsize=(12,6))

sns.barplot(
    data=plot_data,
    x="Language",
    y="Count",
    hue="Category"
)

plt.title("Programming Languages: Current vs Desired")
plt.xlabel("Programming Language")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
remote_by_region = pd.crosstab(
    df["Country"],
    df["RemoteWork"],
    normalize="index"
) * 100

remote_by_region.head()
```

```python
top_countries = df["Country"].value_counts().head(15).index

remote_top_countries = remote_by_region.loc[top_countries]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(11, 8))

sns.heatmap(
    remote_top_countries,
    annot=True,
    fmt=".1f",
    cmap="Blues"
)

plt.title("Remote Work Distribution by Country")
plt.xlabel("Remote Work Type")
plt.ylabel("Country")

plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
import pandas as pd

df["YearsCodePro"] = df["YearsCodePro"].replace({
    "Less than 1 year": 0.5,
    "More than 50 years": 51
})

df["YearsCodePro"] = pd.to_numeric(
    df["YearsCodePro"],
    errors="coerce"
)

df["JobSat"] = pd.to_numeric(
    df["JobSat"],
    errors="coerce"
)
```

```python
pip install scipy
```

```python
correlation_data = df[
    ["JobSat", "YearsCodePro"]
].dropna()

pearson_corr = correlation_data["JobSat"].corr(
    correlation_data["YearsCodePro"],
    method="pearson"
)

spearman_corr = correlation_data["JobSat"].corr(
    correlation_data["YearsCodePro"],
    method="spearman"
)

print(f"Pearson correlation: {pearson_corr:.3f}")
print(f"Spearman correlation: {spearman_corr:.3f}")
```

```python
def interpret_correlation(correlation):
    absolute_corr = abs(correlation)

    if absolute_corr < 0.10:
        strength = "very weak or negligible"
    elif absolute_corr < 0.30:
        strength = "weak"
    elif absolute_corr < 0.50:
        strength = "moderate"
    elif absolute_corr < 0.70:
        strength = "strong"
    else:
        strength = "very strong"

    direction = "positive" if correlation > 0 else "negative"

    return f"{strength} {direction} correlation"


print("Pearson:", interpret_correlation(pearson_corr))
print("Spearman:", interpret_correlation(spearman_corr))
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
employment_education = pd.crosstab(
    df["Employment"],
    df["EdLevel"]
)

employment_education
```

```python
employment_education_pct = pd.crosstab(
    df["Employment"],
    df["EdLevel"],
    normalize="index"
) * 100

employment_education_pct
```

```python
import matplotlib.pyplot as plt

employment_education_pct.plot(
    kind="bar",
    stacked=True,
    figsize=(12, 7)
)

plt.title("Education Level by Employment Status")
plt.xlabel("Employment Status")
plt.ylabel("Percentage of Respondents")
plt.legend(
    title="Education Level",
    bbox_to_anchor=(1.05, 1),
    loc="upper left"
)

plt.tight_layout()
plt.show()
```
