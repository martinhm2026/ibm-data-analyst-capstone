### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
df_null = df.isnull().sum()
df_null
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
import seaborn as sns
import matplotlib.pyplot as plt

# Tomar una muestra representativa del dataset
df_sample = df.sample(500, random_state=42)

# Crear heatmap de valores faltantes
plt.figure(figsize=(12,6))

sns.heatmap(
    df_sample.isnull(),
    cbar=False
)

plt.title("Missing Values Heatmap (Sample of 500 rows)")
plt.xlabel("Columns")
plt.ylabel("Rows")

plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
# Contar valores faltantes en la columna Employment

missing_employment = df['Employment'].isnull().sum()

print("Missing values in Employment:", missing_employment)
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
most_frequent_employment = df['Employment'].mode()[0]

print("Most frequent Employment value:", most_frequent_employment)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Reemplazar valores faltantes de Employment con el valor más frecuente

df['Employment'] = df['Employment'].fillna(
    most_frequent_employment
)
```

```python
# Contar valores faltantes después de la imputación

missing_after = df['Employment'].isnull().sum()

print("Missing values after imputation:", missing_after)
```
