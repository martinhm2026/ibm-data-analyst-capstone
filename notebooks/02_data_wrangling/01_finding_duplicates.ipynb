### Step 4: Removing Duplicate Rows


**Task 2: Remove Duplicates**
   1. Remove duplicate rows from the dataset using the drop_duplicates() function.
2. Verify the removal by counting the number of duplicate rows after removal .


```python
## Write your code here
# Eliminar filas duplicadas
df = df.drop_duplicates()
```

### Step 5: Handling Missing Values


**Task 3: Identify and Handle Missing Values**
   1. Identify missing values for all columns in the dataset.
   2. Choose a column with significant missing values (e.g., EdLevel) and impute with the most frequent value.


```python
# Contar valores faltantes por columna
missing_values = df.isnull().sum()

# Filtrar columnas donde existen valores faltantes
missing_values_df = (
    missing_values[missing_values > 0]
    .to_frame(name='Missing_Count')
)

print(missing_values_df)
```

```python
# Columnas categóricas con valores faltantes
categorical_missing = df.select_dtypes(include='object').columns[
    df.select_dtypes(include='object').isnull().any()
]

print(categorical_missing)
```

```python
# Imputar valores faltantes usando la moda para columnas categóricas

for col in categorical_missing:
    
    most_frequent = df[col].mode()[0]
    
    df[col] = df[col].fillna(most_frequent)
    
    print(f"{col} imputada con: {most_frequent}")
```

```python
df[categorical_missing].isnull().sum()
```

```python
numeric_missing = df.select_dtypes(
    include=['int64', 'float64']
).columns[
    df.select_dtypes(
        include=['int64', 'float64']
    ).isnull().any()
]
```

```python
numeric_missing = numeric_missing.drop(
    'ConvertedCompYearly'
)
```

```python
# Imputar columnas numéricas con la mediana

for col in numeric_missing:
    df[col] = df[col].fillna(df[col].median())
```

```python
df[numeric_missing].isnull().sum()
```
